# Data Visualization

```python
# Visualize categorical variables
fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(15, 4))
i = 0
for i, col in enumerate(['Gender', 'Blood Type', 'Admission Type']):
    sns.countplot(data=data, x=col, ax=axes[i])
    axes[i].set_title(f"Distribution of {col}")
plt.tight_layout()
plt.show()
```

```python
# Visualize categorical variables
fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(17, 4))
i = 0
for i, col in enumerate(['Medical Condition', 'Medication','Test Results']):
    sns.countplot(data=data, x=col, ax=axes[i])
    axes[i].set_title(f"Distribution of {col}")
plt.tight_layout()
plt.show()
```

```python
# Visualize numerical variables
numerical_cols = ['Age', 'Billing Amount', 'Room Number']
fig, axes = plt.subplots(nrows=1, ncols=len(numerical_cols), figsize=(15, 4))
for i, col in enumerate(numerical_cols):
    sns.histplot(data=data, x=col, kde=True, ax=axes[i])
    axes[i].set_title(f"Distribution of {col}")
plt.tight_layout()
plt.show()
```
